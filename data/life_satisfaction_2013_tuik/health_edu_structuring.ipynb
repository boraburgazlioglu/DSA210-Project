```python
#Following code restructured the cleaned healthcare and education data by changing names of columns, rows and adding the coast_stat variable

import pandas as pd
from google.colab import files

path = "/content/drive/MyDrive/DSA210 Project/data/life_satisfaction_2013_tuik/healthcare_and_education_satisfaction_2013_clean.xlsx"

df = pd.read_excel(path)

#Rename columns
df = df.rename(columns={'satisfied_with_healthcare': 'satisfied_with_healthcare_percent'})
df = df.rename(columns={'satisfied_with_education': 'satisfied_with_education_percent'})

#Lower and strip province names
df['province'] = df['province'].str.lower()
df['province'] = df['province'].str.strip()

#Add coastal stat column according to the specification in the README/Data sources and collection
#Prepared manually via the Turkey political map
coastal_dict = {
    'adana':1,'adiyaman':0,'afyonkarahisar':0,'agri':0,'aksaray':0,
    'amasya':0,'ankara':0,'antalya':1,'ardahan':0,'artvin':1,'aydin':1,
    'balikesir':1,'bartin':1,'batman':0,'bayburt':0,'bilecik':0,'bingol':0,
    'bitlis':0,'bolu':0,'burdur':0,'bursa':1,'canakkale':1,'cankiri':0,
    'corum':0,'denizli':0,'diyarbakir':0,'duzce':1,'edirne':1,'elazig':0,
    'erzincan':0,'erzurum':0,'eskisehir':0,'gaziantep':0,'giresun':1,
    'gumushane':0,'hakkari':0,'hatay':1,'igdir':0,'isparta':0,'istanbul':1,
    'izmir':1,'kahramanmaras':0,'karabuk':0,'karaman':0,'kars':0,'kastamonu':1,
    'kayseri':0,'kilis':0,'kirikkale':0,'kirklareli':1,'kirsehir':0,
    'kocaeli':1,'konya':0,'kutahya':0,'malatya':0,'manisa':0,'mardin':0,
    'mersin':1,'mugla':1,'mus':0,'nevsehir':0,'nigde':0,'ordu':1,'osmaniye':0,
    'rize':1,'sakarya':1,'samsun':1,'siirt':0,'sinop':1,'sivas':0,
    'sanliurfa':0,'sirnak':0,'tekirdag':1,'tokat':0,'trabzon':1,'tunceli':0,
    'usak':0,'van':0,'yalova':1,'yozgat':0,'zonguldak':1
}

df['coast_stat'] = df['province'].map(coastal_dict)

df_final = df[['province', 'satisfied_with_healthcare_percent', 'satisfied_with_education_percent', 'coast_stat']]

df_final.to_csv("health_edu_with_coast_stat.csv", sep=';', index=False)

files.download("health_edu_with_coast_stat.csv")
```
