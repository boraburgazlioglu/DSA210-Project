```python
#The following code was used to merge the 3 datasets happiness_scores, healthcare_and_education_satisfaction and wellbeing_indicators
import pandas as pd
from google.colab import files

life_satisfaction = pd.read_csv("/content/drive/MyDrive/DSA210 Project/data/life_satisfaction_2013_tuik/happiness_scores_with_coast_stat.csv", sep=";")
wellbeing = pd.read_csv("/content/drive/MyDrive/DSA210 Project/data/tuik_cis/wellbeing_indicators.csv", sep=";")
health_edu = pd.read_csv("/content/drive/MyDrive/DSA210 Project/data/life_satisfaction_2013_tuik/health_edu_with_coast_stat.csv", sep=";")

health_edu = health_edu.drop(columns=['coast_stat'], errors='ignore')

df = (life_satisfaction.merge(wellbeing, on="province").merge(health_edu, on="province"))

#Rearrange columns for organization
df = df[['province',
         'coast_stat',
         'happiness_score_percent',
         'employment_rate_percent',
         'average_years_in_school',
         'satisfied_with_education_percent',
         'hospital_beds_per_100k',
         'satisfied_with_healthcare_percent']]

df.to_csv("merged_data.csv", sep=';', index=False)

files.download("merged_data.csv")
```
